```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the image to be tested
test_image = cv2.imread('lena_color_256.tif')

# #Since we know that OpenCV loads an image in BGR format, so we need to convert it into RBG format to be able to display 
# #its true colors. Let us write a small function for that.
```

```python
#Converting to grayscale
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
```

```python
# Displaying the grayscale image
plt.imshow(test_image_gray, cmap='gray')
```

```python
# convert it into RBG format to be able to display its true colors. 
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
#detectMultiscale module of the classifier. This function will return a rectangle with coordinates(x,y,w,h) around the detected face.
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5);

# scalefactor: In a group photo, there may be some faces which are near the camera than others.

# minNeighbors This parameter specifies the number of neighbors a rectangle should have to be called a face

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

```python
#((0,255,0),2)
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
#convert image to RGB and show image
plt.imshow(convertToRGB(test_image))
```
