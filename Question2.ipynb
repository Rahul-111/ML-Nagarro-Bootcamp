```python
df_cases=df.query("unit=='cases'")
```

```python
df_cases_caba=df_cases.query("location=='Argentina-CABA'")
```

```python
df_cases_caba['report_date'].unique()
```

```python
#14 unique days
```

```python
df_cases_cata=df_cases.query("location=='Argentina-Catamarca'")
```

```python
df_cases_cata['report_date'].unique()
```

```python
x=df_cases_caba['report_date'].unique()
```

```python
y=df_cases_cata['report_date'].unique()
```

```python
x==y
```

```python
#for a given date we need to calculate sum of all value
```

```python
grp=df_cases_caba.groupby('report_date')
```

```python
df_date=grp.get_group('2016-03-19')
```

```python
df_date.head()
```

```python
x_value_sum_date_vise=[]
df_date['value'].apply(pd.to_numeric).sum()
```

```python
y=df_date['value'].apply(pd.to_numeric).sum()
x_value_sum_date_vise.append(y)
```

```python
x_value_sum_date_vise_Argentina_CABA=[]
gr=df_cases_caba.groupby('report_date')
for day in x:
    df_date=gr.get_group(day)
    y=df_date['value'].apply(pd.to_numeric).sum()
    x_value_sum_date_vise_Argentina_CABA.append(y)
```

```python
x_value_sum_date_vise_Argentina_CABA
```

```python
x_value_sum_date_vise_Argentina_Cata=[]
gr=df_cases_cata.groupby('report_date')
for day in x:
    df_date=gr.get_group(day)
    y=df_date['value'].apply(pd.to_numeric).sum()
    x_value_sum_date_vise_Argentina_Cata.append(y)
```

```python
x_value_sum_date_vise_Argentina_Cata
```

```python
plt.plot(x,x_value_sum_date_vise_Argentina_CABA)
```

```python
plt.plot(x,x_value_sum_date_vise_Argentina_Cata)
```

```python
plt.figure(figsize=(20,10))
plt.plot(x,x_value_sum_date_vise_Argentina_CABA)
plt.plot(x,x_value_sum_date_vise_Argentina_Cata)
plt.xlabel('Date')
plt.ylabel('Sum of value')
plt.xticks(rotation=90)
plt.title('Distribution of Sum of value over entire duration')
```
