```python
# Affime Transformation
```

```python
img=cv2.imread('../Data/glass.jpg')
```

```python
row,cols,_=img.shape
```

```python
pts1=np.float32([[50,50],[200,50],[50,200]])
pts2=np.float32([[10,100],[200,50],[100,250]])

M=cv2.getAffineTransform(pts1,pts2)
dst=cv2.warpAffine(img,M,(cols,row))
```

```python
plt.imshow(dst)
```

```python
M.shape
```

```python
pts1=np.float32([[75,20],[145,20],[85,200],[140,200]])
pts2=np.float32([[0,0],[300,0],[0,300],[300,300]])

M=cv2.getPerspectiveTransform(pts1,pts2)
dst=cv2.warpPerspective(img,M,(300,300))
```

```python
plt.imshow(dst)
```

```python
M.shape
```

```python
#Translation
```

```python
M=np.float32([[1,0,80],[0,1,70]])
dst=cv2.warpAffine(img,M,(cols,row))
dst.shape
```

```python
plt.imshow(dst)
```

```python
#Rotation
```

```python
cv2.getRotationMatrix2D((cols/2,row/2),45,1)
```

```python
M = cv2.getRotationMatrix2D((cols/2,row/2),90,1)#rotate by 90degree
dst = cv2.warpAffine(img,M,(cols,row))
```

```python
plt.imshow(dst)
```
